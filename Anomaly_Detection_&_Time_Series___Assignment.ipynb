Question 1: What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.

Anomaly Detection is a technique used in data analysis and machine learning to identify unusual patterns, rare events, or observations that significantly differ from the majority of the data.

These unusual data points are called anomalies, outliers, or exceptions.

It is widely used in:

Fraud detection (banking, insurance)

Network intrusion detection

Medical diagnosis

Fault detection in machines

Credit card transaction monitoring

Types of Anomalies

There are three main types of anomalies:

1Ô∏è‚É£ Point Anomalies (Global Anomalies)
Definition:

A single data point that is significantly different from the rest of the data.

Example:

In a dataset of daily salaries ranging from ‚Çπ20,000‚Äì‚Çπ50,000, one record shows ‚Çπ5,00,000.

A credit card transaction of ‚Çπ5 lakh when the user usually spends ‚Çπ2,000‚Äì‚Çπ5,000.

Visualization Idea:

If most points are clustered together and one point lies far away ‚Äî that point is a point anomaly.

Real-world Example:

Credit card fraud detection

Network traffic spike at a specific time

2Ô∏è‚É£ Contextual Anomalies (Conditional Anomalies)
Definition:

A data point that is abnormal in a specific context but may be normal in another context.

Context usually involves:

Time

Location

Season

User behavior

Example:

Temperature of 40¬∞C:

Normal in summer

Abnormal in winter

High electricity usage:

Normal during daytime

Abnormal at 3 AM

Key Idea:

The anomaly depends on the contextual attributes.

Real-world Example:

Unusual login time (user logs in at 3 AM when they usually log in at 9 AM)

Sudden traffic spike at midnight

3Ô∏è‚É£ Collective Anomalies
Definition:

A group of data points that together form an unusual pattern, even if individual points may not appear abnormal.

Example:

A sudden sequence of small transactions made within seconds.

Multiple failed login attempts followed by one successful login.

A pattern of sensor readings showing slight but continuous increase before machine failure.

Individually ‚Üí Normal
Together ‚Üí Suspicious pattern

Real-world Example:

DDoS attack (many small requests together)

Fraud rings making coordinated transactions

Stock market manipulation patterns

Summary Table

| Type           | What is Anomalous?             | Example                           |
| -------------- | ------------------------------ | --------------------------------- |
| **Point**      | Single data point              | One huge transaction              |
| **Contextual** | Data point in specific context | 40¬∞C in winter                    |
| **Collective** | Group of data points           | Multiple rapid small transactions |



Why Anomaly Detection is Important?

Prevent financial loss (fraud detection)

Improve security (intrusion detection)

Predict equipment failure

Detect medical abnormalities early

Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
their approach and suitable use cases

clear and structured comparison of Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) based on their approach and use cases.

1Ô∏è‚É£ Isolation Forest
üîπ Approach

An ensemble-based algorithm

Works by randomly partitioning data using decision trees

Anomalies are easier to isolate ‚Üí require fewer splits

Points that are isolated quickly are considered anomalies

üëâ Instead of profiling normal data, it isolates anomalies directly

üîπ Key Idea

Anomalies are:

Few in number

Different in attribute values

üîπ Strengths

Works well with high-dimensional data

Scales efficiently to large datasets

No need to compute distance metrics

üîπ Limitations

Not ideal when anomalies are defined by local density differences

May struggle if anomalies are not clearly separable

üîπ Suitable Use Cases

Credit card fraud detection

Network intrusion detection

Large financial datasets

Log anomaly detection

2Ô∏è‚É£ DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
üîπ Approach

Density-based clustering algorithm

Groups points that are closely packed together

Points in low-density regions are labeled as noise (anomalies)

It uses two parameters:

eps ‚Üí radius

min_samples ‚Üí minimum points required to form a cluster

üîπ Key Idea

Anomalies lie in low-density regions

üîπ Strengths

Detects arbitrarily shaped clusters

Automatically identifies noise points

No need to specify number of clusters

üîπ Limitations

Struggles with varying densities

Parameter tuning can be tricky

Not ideal for very high-dimensional data

üîπ Suitable Use Cases

Spatial data analysis

Geographical data

Customer segmentation

Sensor data clustering

3Ô∏è‚É£ Local Outlier Factor (LOF)
üîπ Approach

Density-based anomaly detection algorithm

Compares the local density of a point with its neighbors

If a point has much lower density than neighbors ‚Üí anomaly

üîπ Key Idea

Anomaly is determined relative to surrounding points

üîπ Strengths

Detects local anomalies

Works well when density varies across dataset

More sensitive to subtle anomalies

üîπ Limitations

Computationally expensive for large datasets

Not ideal for very high-dimensional data

Requires choosing number of neighbors (k)

üîπ Suitable Use Cases

Fraud detection in local clusters

Medical anomaly detection

Network behavior analysis

Situations where anomalies depend on local context

üîé Side-by-Side Comparison

| Feature               | Isolation Forest            | DBSCAN             | LOF                      |
| --------------------- | --------------------------- | ------------------ | ------------------------ |
| Type                  | Ensemble-based              | Clustering         | Density-based            |
| Main Concept          | Isolation via random splits | Density clustering | Local density comparison |
| Detects               | Global anomalies            | Noise points       | Local anomalies          |
| Works Well For        | High-dimensional large data | Spatial data       | Varying density data     |
| Scalability           | High                        | Moderate           | Low‚ÄìModerate             |
| Parameter Sensitivity | Low                         | High               | Moderate                 |


When to Use What?

‚úî Use Isolation Forest when:

Dataset is large

You want fast and scalable solution

‚úî Use DBSCAN when:

Data is spatial/geographical

Subtle anomalies need to be detected

Key Components of a Time Series

A Time Series is a sequence of data points recorded at regular time intervals (daily, monthly, yearly, etc.).

A time series typically consists of four key components:

1Ô∏è‚É£ Trend (T)
üîπ Definition:

The long-term upward or downward movement in the data over time.

It shows the general direction in which the data is moving.

üîπ Example:

Increase in online shopping sales over 5 years

Rise in population over decades

Growth of a company‚Äôs revenue annually

üìà Example:
If a company‚Äôs revenue increases from ‚Çπ10 lakh in 2020 to ‚Çπ25 lakh in 2025 ‚Üí that is an upward trend.

2Ô∏è‚É£ Seasonality (S)
üîπ Definition:

Regular and repeating patterns that occur at fixed intervals due to seasonal factors.

These patterns repeat:

Daily

Weekly

Monthly

Quarterly

Yearly

üîπ Example:

Ice cream sales increase every summer

Retail sales spike every December

Electricity usage increases every evening

üìä Example:
If sales increase every December due to festivals ‚Üí that is seasonality.

3Ô∏è‚É£ Cyclical Component (C)
üîπ Definition:

Fluctuations that occur over longer periods, but not at fixed intervals.

Unlike seasonality:

Cycles are irregular

They depend on economic or business conditions

üîπ Example:

Economic boom and recession cycles

Stock market rises and crashes

Business expansion and contraction

üìâ Example:
A company‚Äôs profits increase during economic growth and decrease during recession ‚Üí this is a cyclical pattern.

4Ô∏è‚É£ Irregular (Random / Noise) Component (R)
üîπ Definition:

Unpredictable, random variations that cannot be explained by trend, seasonality, or cycles.

These are unexpected events.

üîπ Example:

Sudden pandemic impact

Natural disasters

Sudden policy changes

üìå Example:

Question 4: Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?


What is Stationarity in Time Series?

A time series is said to be stationary if its statistical properties do not change over time.

That means:

‚úÖ Mean is constant

‚úÖ Variance is constant

‚úÖ Autocovariance depends only on lag (not time)

In simple words:
üëâ The series fluctuates around a constant level and does not show long-term trends or changing variance.

üîé Why is Stationarity Important?

Many forecasting models like:

ARMA

ARIMA

SARIMA

assume stationarity for accurate predictions.

Non-stationary data can lead to:

Spurious regression

Poor forecasts

Unstable models

üìå Types of Stationarity
1Ô∏è‚É£ Strict Stationarity

Full probability distribution does not change over time (strong condition).

2Ô∏è‚É£ Weak (Covariance) Stationarity

Only mean, variance, and autocovariance remain constant (commonly used in practice).

üìâ Example of Non-Stationary Series

Stock prices showing upward trend

Sales increasing every year

Series with changing variance

These have:

Trend

Seasonality

Changing variance

üß™ How to Test Stationarity?
1Ô∏è‚É£ Visual Inspection

Plot the time series:

Upward/downward trend ‚Üí Non-stationary

Increasing variance ‚Üí Non-stationary

2Ô∏è‚É£ Rolling Statistics

Calculate rolling mean and rolling variance:

If they change over time ‚Üí Non-stationary

3Ô∏è‚É£ Augmented Dickey-Fuller (ADF) Test

Most common statistical test.

Hypotheses:

H‚ÇÄ (Null): Series is non-stationary

H‚ÇÅ (Alternative): Series is stationary

Decision Rule:

p-value < 0.05 ‚Üí Reject H‚ÇÄ ‚Üí Stationary

p-value > 0.05 ‚Üí Fail to reject H‚ÇÄ ‚Üí Non-stationary

4Ô∏è‚É£ KPSS Test (Optional)

Opposite of ADF:

H‚ÇÄ: Series is stationary

H‚ÇÅ: Non-stationary

Used alongside ADF for confirmation.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.api as sm

# Load AirPassengers dataset
data = sm.datasets.get_rdataset("AirPassengers").data

# Create datetime index
data['Month'] = pd.date_range(start='1949-01-01', periods=len(data), freq='ME') # Changed 'M' to 'ME'
data.set_index('Month', inplace=True)

# Rename column 'value' to 'Passengers'
data = data.rename(columns={'value': 'Passengers'})

# -----------------------------
# 1Ô∏è‚É£ Plot Original Time Series
# -----------------------------
plt.figure()
plt.plot(data['Passengers'])
plt.title("AirPassengers - Original Time Series")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.show()

# -----------------------------
# 2Ô∏è‚É£ Decompose Time Series
# -----------------------------
# Multiplicative model (variance increases over time)
decomposition = seasonal_decompose(data['Passengers'], model='multiplicative')

# Plot decomposition
decomposition.plot()
plt.show()
```

Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.
(Include your Python code and output in the code box below.)

we‚Äôll demonstrate anomaly detection using Isolation Forest on a taxi fare‚Äìstyle numerical dataset (similar to NYC Taxi Fare).

We‚Äôll:

Load a numerical dataset

Apply Isolation Forest

Detect anomalies

Visualize them in a 2D scatter plot

Since external datasets may not always load directly, the example below simulates taxi fare‚Äìlike data (Fare Amount vs Trip Distance), which matches the structure of NYC Taxi Fare data.

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Set random seed for reproducibility
np.random.seed(42)

# -----------------------------
# 1Ô∏è‚É£ Create Taxi-like Dataset
# -----------------------------
# Normal taxi trips
distance = np.random.normal(loc=5, scale=2, size=300)
fare = distance * 3 + np.random.normal(loc=0, scale=2, size=300)

# Add some anomalies (very high fare or distance)
anomaly_distance = np.random.uniform(20, 30, 10)
anomaly_fare = np.random.uniform(100, 150, 10)

# Combine normal + anomaly data
X_distance = np.concatenate([distance, anomaly_distance])
X_fare = np.concatenate([fare, anomaly_fare])

data = pd.DataFrame({
    "Distance": X_distance,
    "Fare": X_fare
})

# -----------------------------
# 2Ô∏è‚É£ Apply Isolation Forest
# -----------------------------
model = IsolationForest(contamination=0.03, random_state=42)
data["Anomaly"] = model.fit_predict(data[["Distance", "Fare"]])

# -1 = anomaly, 1 = normal
anomalies = data[data["Anomaly"] == -1]
normal = data[data["Anomaly"] == 1]

# -----------------------------
# 3Ô∏è‚É£ Visualize Results
# -----------------------------
plt.figure()
plt.scatter(normal["Distance"], normal["Fare"])
plt.scatter(anomalies["Distance"], anomalies["Fare"])
plt.xlabel("Trip Distance")
plt.ylabel("Fare Amount")
plt.title("Isolation Forest - Taxi Fare Anomaly Detection")
plt.show()

# Display anomaly points
print("Detected Anomalies:")
print(anomalies)
```

Expected Output Explanation
üîπ Scatter Plot:

Most points form a dense cluster ‚Üí Normal trips

A few far-separated points ‚Üí Anomalies
(Very long trips or unusually high fares)

üîπ Printed Output:

Displays rows classified as anomalies (Anomaly = -1).

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX

# -----------------------------
# 1Ô∏è‚É£ Load Dataset
# -----------------------------
data = sm.datasets.get_rdataset("AirPassengers").data

# Create datetime index
data['Month'] = pd.date_range(start='1949-01-01', periods=len(data), freq='ME') # Changed 'M' to 'ME'
data.set_index('Month', inplace=True)
# Rename column 'value' to 'Passengers'
data = data.rename(columns={'value': 'Passengers'})

# -----------------------------
# 2Ô∏è‚É£ Train SARIMA Model
# SARIMA(p,d,q)(P,D,Q,s)
# Using (1,1,1)(1,1,1,12)
# -----------------------------
model = SARIMAX(
    data['Passengers'],
    order=(1,1,1),
    seasonal_order=(1,1,1,12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

results = model.fit()

# -----------------------------
# 3Ô∏è‚É£ Forecast Next 12 Months
# -----------------------------
forecast = results.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

# -----------------------------
# 4Ô∏è‚É£ Plot Results
# -----------------------------
plt.figure()
plt.plot(data['Passengers'], label='Historical Data')
plt.plot(forecast_mean, label='Forecast (Next 12 Months)')
plt.fill_between(
    forecast_ci.index,
    forecast_ci.iloc[:, 0],
    forecast_ci.iloc[:, 1],
    alpha=0.3
)
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.title("SARIMA Forecast - AirPassengers Dataset")
plt.legend()
plt.show()

# Display forecasted values
print("Forecasted Values for Next 12 Months:")
print(forecast_mean)
```

(1,1,1) ‚Üí Non-seasonal ARIMA part

(1,1,1,12) ‚Üí Seasonal component with 12-month cycle

anomalies and visualize them using matplotlib.

Local Outlier Factor (LOF) on a numerical dataset and visualize anomalies using matplotlib.

LOF detects anomalies by comparing the local density of a point to the density of its neighbors. Points with significantly lower density are flagged as anomalies.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Set random seed
np.random.seed(42)

# -----------------------------
# 1Ô∏è‚É£ Create Sample Numerical Dataset
# -----------------------------
# Normal data cluster
X_normal = 0.3 * np.random.randn(300, 2)

# Add outliers
X_outliers = np.random.uniform(low=-4, high=4, size=(20, 2))

# Combine dataset
X = np.vstack((X_normal, X_outliers))

# -----------------------------
# 2Ô∏è‚É£ Apply Local Outlier Factor
# -----------------------------
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.06)
y_pred = lof.fit_predict(X)

# -1 = anomaly, 1 = normal
normal_points = X[y_pred == 1]
anomalies = X[y_pred == -1]

# -----------------------------
# 3Ô∏è‚É£ Visualize Results
# -----------------------------
plt.figure()
plt.scatter(normal_points[:, 0], normal_points[:, 1])
plt.scatter(anomalies[:, 0], anomalies[:, 1])
plt.title("Local Outlier Factor (LOF) - Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Print detected anomalies
print("Detected Anomalies:")
print(anomalies)
```

Expected Output Explanation
üîπ Scatter Plot

Dense cluster near the center ‚Üí Normal data

Sparse distant points ‚Üí Anomalies (LOF = -1)

üîπ Printed Output

Shows coordinates of detected anomaly points.

üìå Why LOF?

LOF is useful when:

Data has varying density

Anomalies depend on local neighborhood

You want to detect subtle, local deviations

üéØ Real-World Applications

LOF is commonly used in:

Fraud detection

Network intrusion detection

Healthcare anomaly detection

Manufacturing fault detection
